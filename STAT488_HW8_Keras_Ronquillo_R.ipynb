```r
install.packages("ISLR2")
install.packages("glmnet")
library(ISLR2)
library(glmnet)
```

```r
covid <- read.csv(file = '/content/covid_preprocessed_v2.csv')
head(covid)
```

```r
covid_rand = covid[sample(1:nrow(covid)), ]
covid_rand
```

```r
nrow(covid_rand)
```

```r
covid_small2 <- covid_rand[1:20000,]
covid_small2
```

```r
nrow(covid_small2)
```

```r
sample <- sample(c(TRUE, FALSE), nrow(covid_small2), replace=TRUE, prob=c(0.7,0.3))
train  <- covid_small2[sample, ]
test   <- covid_small2[!sample, ]
```

```r
train
test
```

```r
nrow(test)
nrow(train)
```

```r
library(keras)
x_train <- model.matrix(CLASIFFICATION_FINAL ~ . -1, data = train)
y_train <- to_categorical(train$CLASIFFICATION_FINAL)
```

```r
x_train
```

```r
y_train
```

```r
x_test <- model.matrix(CLASIFFICATION_FINAL ~ . -1, data = test)
y_test <- to_categorical(test$CLASIFFICATION_FINAL)
```

```r
library(nnet)
check <- nnet::nnet(CLASIFFICATION_FINAL ~ . -1, data = train, size = 25)
predict(check, newdata = test)
```

```r
library(keras)
nn_covid <- keras_model_sequential() %>%

  layer_dense(units = 128, activation = 'relu', input_shape = ncol(x_train)) %>%
  layer_dense(units = 2, activation = 'softmax')


summary(nn_covid)
```

```r
nn_covid %>% compile(loss = "BinaryCrossentropy",
    optimizer = optimizer_rmsprop(),
    metrics = list("accuracy")
   )


history <- nn_covid %>% fit(
  x_train, y_train, epochs = 250, batch_size = 32,
    #x[-testid, ], y[-testid], epochs = 15, batch_size = 32,
    validation_data = list(x_test, y_test)
  )


plot(history)
###
npred <- predict(nn_covid, x_test)
mean(abs(y_test - npred))
```
